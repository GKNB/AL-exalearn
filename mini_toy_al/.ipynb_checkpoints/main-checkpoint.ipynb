```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import random
```

```python
class FullModel(torch.nn.Module):
    def __init__(self):
        super(FullModel, self).__init__()
        self.fc1 = torch.nn.Linear(1, 5)
        self.fc2 = torch.nn.Linear(5, 5)
        self.fc3 = torch.nn.Linear(5, 1)
        self.relu = torch.nn.ReLU()
        
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        return x
```

```python
def func(x):
    return np.power(x, 1) * np.power(np.sin(1.0 / 3.0 * x), 2)
```

```python
def avg_l2_diff(y1, y2):
    return np.average(np.power(y1-y2, 2))
```

```python
seed = 1
```

```python
torch.manual_seed(seed)
random.seed(seed)
np.random.seed(seed)
torch.use_deterministic_algorithms(True)
```

```python
model = FullModel()
```

```python
x_train = np.arange(-2.4, 3.2, 0.8)
#x_train = np.arange(-2.4, 2.8, 0.4)
#x_train = np.arange(-2.4, 2.6, 0.2)
#x_train = np.arange(-2.4, 2.5, 0.1)
```

```python
y_train = func(x_train)
print(y_train)
```

```python
x_train_torch = torch.from_numpy(x_train).float()
y_train_torch = torch.from_numpy(y_train).float()
train_dataset = torch.utils.data.TensorDataset(x_train_torch, y_train_torch)
train_loader = torch.utils.data.DataLoader(
    train_dataset, shuffle=True)
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, weight_decay=0.001)
```

```python
criterion = torch.nn.MSELoss()
```

```python
def train(epoch):
    model.train()
    
    running_loss  = torch.tensor(0.0)

    for batch_idx, current_batch in enumerate(train_loader):     
        inp, current_batch_y = current_batch[0],        current_batch[1]

        optimizer.zero_grad()
        output = model(inp)
        gndtruth = current_batch_y

        loss = criterion(output, gndtruth)
        loss.backward()
        optimizer.step()
        running_loss  += loss.item()

    running_loss  = running_loss  / len(train_loader)
    print("Average loss: {:15.8f}".format(running_loss))
```

```python
for epoch in range(1, 1000):
    train(epoch)
```

```python
x_pred_np = np.arange(-2.5, 2.5, 0.0001).reshape(-1, 1)
x_pred = torch.from_numpy(x_pred_np).float()
```

```python
y_true = func(x_pred)
y_pred = model(x_pred)
```

```python
x_np = x_pred.numpy().reshape(-1)
y_pred_np = y_pred.detach().numpy().reshape(-1)
y_true_np = y_true.numpy().reshape(-1)
```

```python
print(avg_l2_diff(y_pred_np, y_true_np))
```

```python
plt.figure(figsize=(8, 6))

# Plot y1 and y2 against x
plt.plot(x_np, y_pred_np, label='pred', marker='o')
plt.plot(x_np, y_true_np, label='true', marker='s')

# Set labels and title
plt.xlabel('x')
plt.ylabel('Values')
plt.title('Plotting PyTorch Tensors')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Plot y1 and y2 against x
plt.plot(x_np, np.power(y_pred_np - y_true_np, 2), label='l2-diff', marker='o')

# Set labels and title
plt.xlabel('x')
plt.ylabel('Values')
plt.title('Plotting PyTorch Tensors l2-difference')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
